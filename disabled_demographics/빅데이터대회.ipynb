## 년도, 월, 지역별로 원하는대로 df만들기

```python
def numOfRegistBy_disabilityType_year_month_region(year, month, region):
    yeardf = dfs[years[year]]
    
    lst = []
    for i in range(0, len(yeardf)):
        lst.append(yeardf['통계년월'].loc[i])
        
        if month == lst[i][4:]:
            monthdf = yeardf[yeardf['통계년월']==lst[i]]
    
    regiondf = monthdf[monthdf['통계시도명']==region]
    
    
    return regiondf
```
